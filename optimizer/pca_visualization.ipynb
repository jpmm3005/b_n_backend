```python
import numpy as np
import matplotlib.pyplot as plt

# X is the transformed data (A @ P)
X = np.array([
    [1508.5, 0.67661],
    [1338.7, 5.6226],
    [1588.4, -4.0683],
    [1218.8, -3.7342]
])

# Optional: assign cluster labels (manually or from clustering)
labels = ['Cluster 1', 'Cluster 2', 'Cluster 3', 'Cluster 4']

# Plot
plt.figure(figsize=(8, 6))
for i in range(len(X)):
    plt.scatter(X[i, 0], X[i, 1], s=100, label=labels[i])
    plt.text(X[i, 0]+10, X[i, 1], f'{labels[i]}', fontsize=10)

plt.title("$X = AP$")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.grid(True)
plt.legend()
plt.show()

```
